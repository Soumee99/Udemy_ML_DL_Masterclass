* three main tyopes of distribution plots:
 1. Rug plot

### distribution within categiries
* Boxplot : displays the distribution of a continuous variable.
* violinplot
* Swarmplot
* boxenplot

### For boxplot
 * Q1: 1st quartile i.e. 25 5 points are below 25%
 * Q3: 3rd quartile i.e. 25 % of all points are above Q3
 * the whiskers are defined by 1.5 X IQR
 * Anything otside the q1 and q3 range is outliers
 
 

### Viloin plot

* The violin plot plays a similar role as the box plot
 * it displays the probability density across the data using a KDE
 * it is a kde plot mirrored and joined
 

### Swarmplot

* It is very simple and simply shows all the data points in the distribution. For large datasets it will display general distribution of them.
* it also shows the number of counts per category which the box plot and violin plot failed to do

```python
sns.PairGrid(df)
# an initial outline o  grids are created and it is waiting for the data to get mapped
```

```python
#Customisation of pairplot is pairgrid
g=sns.PairGrid(df,hue='gender')

g=g.map_upper(sns.scatterplot) # for upper diagonal

g=g.map_diag(sns.kdeplot)

g=g.map_lower(sns.kdeplot) # for lower diagonal

g=g.add_legend()
```

### Matrix plots

the visual equivalent of displaying a pivot table
* The matrix plot displays all the data passed in, visualizing all the numeric values in a DataFrame

Two main types:
* heatmap
* clustermap

```python
df=pd.read_csv("C:\\Users\\2211444\Desktop\\Udemy Python Masterclass\\UNZIP_FOR_NOTEBOOKS_FINAL\\05-Seaborn\\country_table.csv")
```

```python
sns.heatmap(df.drop('Life expectancy',axis=1),linewidth=0.5,annot=True)
```
