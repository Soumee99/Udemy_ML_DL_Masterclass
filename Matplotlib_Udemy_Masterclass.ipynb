### Object Oriented Programming with matplotlib

```python
#plt.figure creates a blank canvas object
#inside this figure we need to add axes
# axes=fig.add_axes([x,y,w,h]) w==width,h==height
#axes.plot(x,y)
```

```python
a=np.linspace(0,1,11)
b=a**4
```

```python
a,b
```

```python
x=np.arange(0,10)
```

```python
y=x*2
```

```python
x,y
```

```python
fig=plt.figure() #creating a blank canvas

axes=fig.add_axes([0,0,1,1]) #adding axes to fig object

axes.plot(a,b) #plotting 

plt.show()
```

```python
fig=plt.figure() #creating a blank canvas

#Large axes
axes1=fig.add_axes([0,0,1,1]) #adding axes to fig object
axes1.plot(a,b)

axes1.set_xlim(0,8)
axes1.set_ylim(0,600)
axes1.set_xlabel('A')
axes1.set_ylabel("B")
axes1.set_title("Power of 4")

#small axes
axes2=fig.add_axes([0.2,0.2,0.25,0.25])
axes2.plot(x,y)
axes2.set_xlim(1,2)
axes2.set_ylim(0,50)
axes2.set_xlabel('A')
axes2.set_ylabel("B")
axes2.set_title("Zoomed in")




plt.show()
```

```python
fig=plt.figure(figsize=(2,2),dpi=200)

axes1= fig.add_axes([0,0,1,1])

axes1.plot(a,b)

fig.savefig("new_figure.png",bbox_inches='tight') #inorder to have a bounding box around my figure
```

### Subplots

#### fig,axes=plt.subplots(nrows=1,ncols=2)
* plt.subplots() returns a tuple which by common convention we label (fig,axes):
* fig : this is the entire gfigure canvas
* axes: this is a numpy array holding each of the axes acording to position in the overall canvas

```python
a,b
```

```python
x,y
```

```python
fig,axes=plt.subplots(nrows=1,ncols=2)

axes[0].plot(x,y)
axes[1].plot(a,b)
```

```python
type(axes)
```

```python
fig,axes=plt.subplots(nrows=2,ncols=2)

axes[0][0].plot(x,y)
axes[1][1].plot(a,b)
```

```python
fig,axes=plt.subplots(nrows=3,ncols=1)

axes[0].plot(x,y)
plt.tight_layout() #so no overlapping is there
```

```python
fig,axes=plt.subplots(nrows=2,ncols=2)

axes[0][0].plot(x,y)
axes[0][0].set_title("1st plot")
axes[0][1].plot(x,y)
axes[0][1].set_title("2nd plot")
axes[1][0].plot(x,y)
axes[1][0].set_title("3rd plot")
axes[1][1].plot(a,b)
axes[1][1].set_title("4th plot")

fig.suptitle("figure level",fontsize=16) # assiggning title to the figure as a whole

plt.tight_layout()
fig.subplots_adjust(wspace=0.9)#for increasing space between subplots

```

## Adding legend  & STYLING

```python
fig=plt.figure(figsize=(2,2),dpi=100)

axes1= fig.add_axes([0,0,1,1])

axes1.plot(a,b,label='a vs b',color="red",lw=3,ls="--",marker="o",markersize=5)
axes1.plot(a*2,b*2,label="a*2 vs b*2",color="green",lw=4,ls="--",marker="*",ms=6)
# we can also add HEX code color , search the code of a HEX color on google
#markerfacecolor,markeredgewidth,markeredgecolor can also be included
axes1.legend(loc=(0.2,0.5))
```
